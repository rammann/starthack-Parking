```python
import json
import matplotlib.pyplot as plt
```

```python
with open('data.json', 'r') as file:
    data = json.load(file)
```

```python
days = [0]*31
```

```python
for entry in data:
    date = entry["fields"]["created"][5:10]
    day = int(date[3:])
    print(day)
    days[day-1] += 1
```

```python
print(days)
```

```python
plt.plot(days)
```
